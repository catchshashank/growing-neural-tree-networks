```
# Root Node
def create_softmax_node(input_dim=784, num_classes=10, lr=1e-3):
    # Root (and children) are "one-node" classifiers: no hidden layer
    model = Sequential([
        Dense(num_classes, activation="softmax", input_shape=(input_dim,))
    ])
    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model
```

```
def ce_loss_from_probs(probs, y_true_oh):
    eps = 1e-12
    probs = np.clip(probs, eps, 1.0 - eps)
    return float(-np.mean(np.sum(y_true_oh * np.log(probs), axis=1)))

def probs(model, X_batch):
    return model.predict(X_batch, verbose=0)

def pred_int_from_probs(p):
    return np.argmax(p, axis=1)
```
